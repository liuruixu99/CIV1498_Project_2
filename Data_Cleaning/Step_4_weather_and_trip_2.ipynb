```python
df = pd.read_csv('/content/drive/MyDrive/CIV1498 Final Project/Data/cleaned_data3/period_bike_and_weather.csv')
```

```python
import datetime
from datetime import date                                          
import holidays

time = []
weekday = []
ca_holidays = holidays.Canada()    
holiday_lst = [] 
weekend_lst = []                    


for idx, row in df.iterrows():
  date = datetime.datetime(row['Year'], row['Month'], row['Day'])
  time.append(date)
  weekday.append(date.weekday())

  #check whether the date is holiday
  holiday_name = ca_holidays.get(date)
  if holiday_name != None: 
    holiday_lst.append(True)
  else:
    holiday_lst.append(False)


df['Time'] = time
df['Weekday'] = weekday
df['is_holiday'] = holiday_lst
```

```python
is_weekend = (df['Weekday'] == 5) | (df['Weekday'] == 6)
df['is_weekend'] = is_weekend
```

```python
header = ['Time', 'Year', 'Month', 'Day', 'period', 'Weekday', 'is_holiday', 'is_weekend', 'Temp (Â°C)',
       'Dew Point Temp (Â°C)', 'Rel Hum (%)', 'Wind Dir (10s deg)',
       'Wind Spd (km/h)', 'Visibility (km)', 'Stn Press (kPa)', 'Weather',
       'total trips', 'total member', 'total casual', 'mean duration']
df = df[header]
df
```
